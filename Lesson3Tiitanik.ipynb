```python
import pandas as pd
```

```python
train_org = pd.read_csv("data/titanic/train.csv")
```

```python
# train_org.drop("Name", axis=1)
df = train_org.drop(columns = ["Name", "PassengerId", "Ticket", "Cabin"], axis=1)
```

```python
df
```

```python
df = df.dropna(subset=['Age', 'Embarked'])
df = df.reset_index(drop=True)
df = df[df['Fare'] != 0]
df = df.reset_index(drop=True)
```

```python
df
```

```python
df["Male"] = (df["Sex"] == "male").astype(int)
```

```python
df
```

```python
df["EmbarkedS"] = (df["Embarked"] == "S").astype(int)
```

```python
df
```

```python
df["EmbarkedC"] = (df["Embarked"] == "C").astype(int)
```

```python
df
```

```python
df["Pclass_1"] = (df["Pclass"] == 1).astype(int)
df["Pclass_2"] = (df["Pclass"] == 2).astype(int)
```

```python
df
```

```python
max_age = df["Age"].max()
max_age
df["AgeN"] = df["Age"] / max_age
df
```

```python
import numpy as np
df['log_Fare'] = np.log(df['Fare'])
df
```
