# Lets try with 2 liner functions and summing them
`f(x) = ax + bx`

First we create random a and b

```python
a = torch.rand(10) - 0.5
a.requires_grad_()
b = torch.rand(10) - 0.5
b.requires_grad_()
print(a)
print(b)
```

Then we create function `f(x) = a*x + b*x`

```python
def f2(a,b,x):
    return a*x + b*x

def make_fx2(a, b):
    return partial(f2, a, b)

def fx_mae2(a, b):
    f2 = make_fx2(a, b)
    return mae(f2(x), y)
```

Test manually

```python
loss = fx_mae2(a,b)
loss.backward()
print(loss)
print(a.grad)
print(b.grad)
```

Test in loop

```python
for i in range(10):
    loss = fx_mae2(a,b)
    loss.backward()
    with torch.no_grad(): 
        a -= a.grad*0.01
        b -= b.grad*0.01
    # if i % 10 == 0:
    print(f'step={i}; loss={loss:.2f} ')
    # print(a.grad)
    # print(a)
```

Lets try with 5 params

```python
a = torch.rand(10) - 0.5
a.requires_grad_()
b = torch.rand(10) - 0.5
b.requires_grad_()
c = torch.rand(10) - 0.5
c.requires_grad_()
d = torch.rand(10) - 0.5
d.requires_grad_()
e = torch.rand(10) - 0.5
e.requires_grad_()


def f5(a,b,c,d,e,x):
    return a*x + b*x + c*x + d*x + e*x

def make_fx5(a, b, c, d, e):
    return partial(f5, a, b, c, d, e)

def fx_mae5(a, b, c, d, e):
    f5 = make_fx5(a, b, c, d, e)
    return mae(f5(x), y)
```

```python
for i in range(10):
    loss = fx_mae5(a,b,c,d,e)
    loss.backward()
    with torch.no_grad(): 
        a -= a.grad*0.001
        b -= b.grad*0.001
        c -= c.grad*0.001
        d -= d.grad*0.001
        e -= e.grad*0.001
    # if i % 10 == 0:
    print(f'step={i}; loss={loss:.2f} ')
    # print(a.grad)
    # print(a)
```

Lets try with n params

```python
an = []
for i in range(10000):
    a = torch.rand(10) - 0.5
    # a = a.unsqueeze(1)
    a.requires_grad_()
    an.append(a)

def f_n(an,x):
    val = sum(value * x for value in an)
    mask = val < 0
    val[mask] = 0
    # print(val)
    # if val < 0:
    #     val = 0
    return val

def make_fx_n(an):
    return partial(f_n, an)

def fx_mae_n(an):
    fn = make_fx_n(an)
    return mae(fn(x), y)

# for a in an:
#     print(a)
```

```python
loss = fx_mae_n(an)
loss.backward()
print(loss)
print(an[0].grad)
print(an[1].grad)
# for a in an:
#     print(a)
```

```python
for i in range(10):
    loss = fx_mae_n(an)
    loss.backward()
    with torch.no_grad(): 
        for value in an:
            value -= value.grad*0.001 
            # print(value)
    print(f'step={i}; loss={loss:.2f} ')
```
