## Reading and Writting Data in Text Formats

```python
import pandas as pd
import numpy as np
```

```python
chunker = pd.read_csv("../data/book_pfda/ex6.csv", chunksize = 1000)
```

```python
tot = pd.Series([], dtype = np.int64)
for piece in chunker:
    tot = tot.add(piece["key"].value_counts(), fill_value=0)

tot = tot.sort_values(ascending=False)
tot[:10]
```

## Interacting with Web APIs

```python
import requests
```

```python
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
resp = requests.get(url)
resp.raise_for_status()
resp
```

```python
data = resp.json()
```

```python
data[0]["title"]
# type(data[0])
```

```python
# We can pass data to DataFrame
issues = pd.DataFrame(data, columns=["number", "title", "labels", "state"])
issues
```
