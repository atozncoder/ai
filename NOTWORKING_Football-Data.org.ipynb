```python
import requests
import pandas as pd
from pathlib import Path
import os
```

```python
data_path = Path("./ignore/footbal_data_org")
os.makedirs(data_path, exist_ok = True)

base_url = "http://api.football-data.org"
```

```python
response = requests.get(f"{base_url}/v4/competitions/")
competitions_json = response.json()
```

```python
# type(competitions_json['competitions'])
# competitions_json['competitions']
competitions = pd.json_normalize(competitions_json['competitions'])
competitions
```

```python
competitions.columns
```

```python
competitions.to_csv(data_path/"competitions.csv")
```
